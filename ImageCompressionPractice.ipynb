```python
import matplotlib.pyplot as plt
from matplotlib.image import imread
import numpy as np
```

```python
from google.colab import files
uploaded = files.upload()
image_path = next(iter(uploaded))
```

```python
color_image = imread(image_path)
color_image.shape
```

```python
color_image = np.mean(color_image, -1)
plt.imshow(color_image)
```

```python
plt.set_cmap('gray')
plt.imshow(color_image)
grayscale = color_image
```

```python
A = grayscale
U, S, Vt = np.linalg.svd(A, full_matrices = False)
S = np.diag(S)
print(U.shape, S.shape, Vt.shape)
```

```python
def compress(r):
  approx = U[:, :r] @ S[:r, :r] @ Vt[:r, :]
  approx_img = plt.imshow(approx)
  return approx_img

compressed_Img = compress(5)
plt.set_cmap('gray')
plt.show()
```
